```python
import numpy as np
import pandas as pd
import re
import nltk
```

```python
df_in = pd.read_csv('Data/inbounds_data.csv')
#df_out = pd.read_csv('Data/outbounds_data.csv')
df_all = pd.read_csv('Data/all_data.csv')
```

```python
#First inbound from users
first_inbound = df_all[pd.isnull(df_all.in_response_to_tweet_id)]
in_out = pd.merge(first_inbound, df_all, left_on='tweet_id', right_on='in_response_to_tweet_id').sample(frac=1)

# Outbounds replies from companies
in_out = in_out[in_out.inbound_y ^ True]
in_out.tail()
```

```python
#Drop all unecessary features
in_out = in_out.drop(['tweet_id_x', 'inbound_x', 'created_at_x','response_tweet_id_x', 'in_response_to_tweet_id_x', 
          'tweet_id_y', 'inbound_y','created_at_y', 'response_tweet_id_y', 'in_response_to_tweet_id_y'], axis = 1)
in_out.shape
```

```python
in_out.tail()
```

```python
# Get Apple and Amazon data separately
apple = in_out[(in_out['author_id_y'] == 'AppleSupport' )]
amazon = in_out[(in_out['author_id_y'] == 'AmazonHelp' )]

apple = apple.drop(['author_id_x', 'author_id_y'], axis = 1)
amazon = amazon.drop(['author_id_x', 'author_id_y'], axis = 1)
apple.tail()
```

```python
apple.to_csv (r'\Data\text_apple.csv', index = False, header=True)
amazon.to_csv (r'\Data\text_amazon.csv', index = False, header=True)
```
